```python
import pandas as pd



df = pd.read_csv("Incidents.csv")

df.head()
```

```python
columns_to_keep = ["Incident name",	"Brief description of the incident",	"What is the motive? Why is the state cracking down? (if known)", "Country where incident took place",	"Date of incident"]
clean_df = df[columns_to_keep]
year_of_incident = []
for row in clean_df["Date of incident"]:
    row_str = str(row)
    if row_str[-4:].isdigit():
        year = row_str[-4:]
        year_of_incident.append(year)
    else:
        year_of_incident.append(None)
        
clean_df["Year of incident"] = year_of_incident
clean_df.head(67)
```

```python



country_replacements = {
    "China (and Hong Kong)": "China",
    "Russia,Ukraine": "Russia, Ukraine",
    "United States of America,China (and Hong Kong)": "United States, China"  
 
}



clean_df["Country where incident took place"] = clean_df["Country where incident took place"].replace(country_replacements)


new_rows = []
for _, row in clean_df.iterrows():
    countries = [c.strip() for c in row["Country where incident took place"].split(",")]
    for country in countries:
        new_row = row.copy()
        new_row["Country where incident took place"] = country
        new_rows.append(new_row)

clean_df = pd.DataFrame(new_rows)


clean_df.to_csv("cleaned_incidents.csv", index=False)

```

```python
turkish_to_english_months = {
    "Ocak": "January",
    "Şubat": "February",
    "Mart": "March",
    "Nisan": "April",
    "Mayıs": "May",
    "Haziran": "June",
    "Temmuz": "July",
    "Ağustos": "August",
    "Eylül": "September",
    "Ekim": "October",
    "Kasım": "November",
    "Aralık": "December"
}

def convert_turkish_date_to_english(date_str):
    if isinstance(date_str, str):  
        for tr, en in turkish_to_english_months.items():
            if tr in date_str:
                return date_str.replace(tr, en)
    return date_str  


clean_df["Date of incident"] = clean_df["Date of incident"].apply(convert_turkish_date_to_english)


clean_df

```
